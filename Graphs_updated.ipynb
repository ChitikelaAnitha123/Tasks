```python
#1. Titles and Labels:



import matplotlib.pyplot as plt
import numpy as np



x = np.linspace(10, 100, 1000)
y = np.sin(x)



plt.plot(x, y)
plt.title("Sine Curve")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#2. Legends:



import matplotlib.pyplot as plt
import numpy as np



x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)



plt.plot(x, y1, label="Sine")
plt.plot(x, y2, label="Cosine")
plt.legend()
plt.show()
```

```python
#Colors:



import matplotlib.pyplot as plt
import numpy as np



x = np.linspace(0, 10, 100)
y = np.sin(x)



plt.plot(x, y, color="red")
plt.show()

```

```python
#Line Styles and Markers:



import matplotlib.pyplot as plt
import numpy as np



x = np.linspace(0, 10, 100)
y = np.sin(x)



plt.plot(x, y, linestyle="-", marker=".")
plt.show()
```

```python
#Grid and Axes:



import matplotlib.pyplot as plt
import numpy as np



x = np.linspace(0, 10, 100)
y = np.sin(x)



plt.plot(x, y)
plt.grid(True)
plt.show()

```

```python
#Figure Size and DPI:



import matplotlib.pyplot as plt
import numpy as np



x = np.linspace(0, 10, 100)
y = np.sin(x)



plt.figure(figsize=(8, 6), dpi=100)
plt.plot(x, y)
plt.show()
```

```python
#ARROW Labels:



import matplotlib.pyplot as plt
import numpy as np



x = np.linspace(0, 10, 100)
y = np.sin(x)



plt.plot(x, y)
plt.annotate("Maximum", xy=(np.pi/2, 1), xytext=(np.pi, 0.5),
             arrowprops=dict(arrowstyle="->"))
plt.text(7, 0.5, "Text Example")
plt.show()


```
