```python
#What might be an interesting thing to do, is aside from using regression modelling, is to set an arbitrary cutoff for your dependent variable (wine quality) at e.g. 7 or higher getting classified as 'good/1' and the remainder as 'not good/0'
```

```python
wine_df['quality']=wine_df.quality.apply(lambda x:1 if x>7 else 0)
```

```python
wine_df['quality'].value_counts()
```

```python
wine_df
```

```python
y=wine_df['quality']
X=wine_df.drop(['quality'],axis=1)
```

```python
#spliting data set into tarin and test data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print("X_train",X_train.shape)
print("X_test",X_test.shape)
print("y_train",y_train.shape)
print("y_test",y_test.shape)
```

```python
#model training over dataset
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,classification_report
from sklearn.metrics import precision_recall_curve, auc, confusion_matrix,ConfusionMatrixDisplay
```

```python
# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
print(classification_report(y_test,y_pred))
```

```python
style.use('classic')
cm=confusion_matrix(y_test,y_pred,labels=model.classes_)
```

```python
disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
```

```python
disp.plot()
print("TN: ",cm[0][0])
print("FN: ",cm[1][0])
print("TP: ",cm[1][1])
print("FP: ",cm[0][1])
```

```python
# validating model
```

```python
# Doing cross validation on our model
```

```python
from sklearn.model_selection import cross_val_score, KFold


```

```python
num_folds = 5
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)
```

```python
scores = cross_val_score(model, X, y, cv=num_folds, scoring='accuracy')
```

```python
for fold_idx, score in enumerate(scores, start=0):
    print(f"Fold {fold_idx}: Accuracy = {score:.2f}")
```

```python
#Doing hyper parameter tunning 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define hyperparameters for grid search
param_grid = {
    'C': [0.01, 0.1, 1, 10]
}
# Perform grid search with cross-validation
grid_search = GridSearchCV(estimator=model , param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
# Get the best hyperparameters from the grid search
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```
