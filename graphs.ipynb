```python
import matplotlib.pyplot as plt



# Sample data

x = [5,10,15,20]

y = [10,4, 6, 14]
x1 = [1,2,3,4]

y1 = [4,25,26,27]



# Create a line plot

plt.plot(x, y,x1,y1)




# Add labels and title

plt.xlabel('X-axis label')

plt.ylabel('Y-axis label')

plt.title('Line Plot Example')



# Display the plot

plt.show()
```

```python
import seaborn as sns

import numpy as np

import matplotlib.pyplot as plt



# Generate sample data (correlation matrix)

data = np.random.rand(5, 5)



# Create a heatmap using Seaborn

sns.heatmap(data, annot=True, cmap='coolwarm')



# Add title

plt.title('Heatmap Example')



# Display the plot

plt.show()
```

```python
import matplotlib.pyplot as plt



# Sample data

categories = ['Category A', 'Category B', 'Category C', 'Category D']

values = [30, 20, 15, 35]



# Create a pie chart using Matplotlib

plt.pie(values, labels=categories, autopct='%1.3f%%', startangle=90)



# Add title

plt.title('Pie Chart Example')



# Display the plot

plt.show()
```

```python




import plotly.graph_objects as go



# Sample data

x = [1, 2, 3, 4, 5]

y = [10, 8, 12, 6, 14]

z = [5, 9, 3, 7, 2]



# Create a 3D scatter plot using Plotly

fig = go.Figure(data=go.Scatter3d(x=x, y=y, z=z, mode='markers'))



# Add title

fig.update_layout(title='3D Scatter Plot Example')



# Show the plot

fig.show()

```

```python
import matplotlib.pyplot as plt



# Sample data

years = [2010, 2011, 2012, 2013, 2014]

series_a = [20, 25, 30, 35, 40]

series_b = [15, 18, 25, 28, 30]



# Create an area plot using Matplotlib

plt.stackplot(years, series_a, series_b, labels=['Series A', 'Series B'], alpha=0.8)



# Add labels and title

plt.xlabel('Years')

plt.ylabel('Values')

plt.title('Area Plot Example')

plt.legend(loc='upper right')



# Display the plot

plt.show()
```

```python
import matplotlib.pyplot as plt



# Sample data

data = [np.random.normal(10, 2, 100), np.random.normal(15, 3, 100), np.random.normal(12, 1, 100)]



# Create a box plot using Matplotlib

plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])



# Add title

plt.title('Box Plot Example')



# Display the plot

plt.show()

```

```python
import seaborn as sns

import matplotlib.pyplot as plt



# Sample data

df = sns.load_dataset('iris')



# Create a pair plot using Seaborn

sns.pairplot(df, hue='species')



# Add title

plt.title('Pair Plot Example')



# Display the plot

plt.show()
```

```python
from wordcloud import WordCloud

import matplotlib.pyplot as plt



# Sample text

text = "Data visualization in Python is powerful and effective. Python provides various libraries for creating impressive visualizations."



# Create a word cloud using WordCloud

wordcloud = WordCloud().generate(text)



# Display the generated image

plt.imshow(wordcloud, interpolation='bilinear')

plt.axis('off')

plt.title('Word Cloud Example')

plt.show()
```
